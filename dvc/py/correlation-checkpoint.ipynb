```python
csv["horsepower"] = pd.to_numeric(csv["horsepower"].replace("?",0.0)) # here we have changed the datatype and also replaced the 
# data wherever the ? existed...
csv.info()
csv.corr()
```

```python
pd.plotting.scatter_matrix(csv, figsize=(15,15), alpha =0.8)
plot.show()
```

```python
pcplot(csv, "cylinders", cols = ["acceleration","mpg","origin"]) # the graph means in csv data, cylinders vs (acc and mpg and origin)
plot.show()
```

```python
pd.crosstab(csv.cylinders , csv["model year"]) # useful to get information or data on production wise...
```

```python
#  outliers = 
# 25% or 25th percentile from df.describe() = is also called quantile 1.
# 75th percentile is also called quantile 3 or Q3
# IQR = inter quantile range = Q3-Q1
# high point = Q3 + (IQR*1.5)
# low point = Q1 - (IQR*1.5)
csv = pd.read_csv("auto-mpg.csv")
q1 = csv.acceleration.quantile(0.25)
q3 = csv.acceleration.quantile(0.75)
iqr = q3-q1
highpt = q3 + (iqr*1.5)
lowpt=q1 - (iqr*1.5)
print(highpt, lowpt)
csv.loc[(csv.acceleration > lowpt)| (csv.acceleration < highpt)]
```

```python
csv_df = csv.loc[(csv.acceleration >=lowpt) & (csv.acceleration <= highpt)]
plot.boxplot(csv.acceleration)
plot.show()
```
